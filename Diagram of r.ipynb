```python
# Import libraries needed:
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.optimize import curve_fit
%matplotlib inline
from scipy.optimize import curve_fit

# Set global font size for plots:
mpl.rcParams['font.size'] = 14
# (see https://matplotlib.org/tutorials/introductory/customizing.html).
```

```python
x = np.array([0,2600,5100,7700,10200,12700,15400,17800,20500,23200,25500])
y = np.array([0.000660952,0.002132197,0.003655066,0.005181209,0.006727481,0.008152247,0.009807668,0.011162397,0.012751433,0.014306747,0.016102212])
dy = np.array([5.16654E-11,6.73939E-11,7.15104E-11,9.57691E-11,1.62534E-10,1.37474E-10,1.98485E-10,2.37765E-10,1.84317E-10,2.52104E-10,2.39808E-10])
```

```python
plt.plot(x, y, 'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.title('R dependence');
```

```python
def function(X,A):
    y = A*X
    return y
fit, cvm = curve_fit(function, x, y, sigma = dy)
err = np.sqrt(np.diag(cvm))
[print("a_{0} = {1:0.10f} \xb1 {2:0.10f}".format(i, fit[i], err[i])) for i in range(len(fit))];
```

```python
dy = np.array([5.16654E-11,6.73939E-11,7.15104E-11,9.57691E-11,1.62534E-10,1.37474E-10,1.98485E-10,2.37765E-10,1.84317E-10,2.52104E-10,2.39808E-10])
xline = np.array(np.linspace(0,30000,10000))
yline = fit[0]*xline
plt.plot(x, y, 'ro',)                    # plot data as red circles.
plt.plot(xline, yline, 'k', label="y=(64.5±0.2)*10^-8")   # plot black line for fit.
plt.errorbar(x,y,yerr=dy,fmt='o',ecolor='hotpink',capsize=6)
plt.legend(loc="best")
plt.xlabel('R/Ω')
plt.ylabel('Voltage square/mV^2')
plt.title('R dependence')
plt.savefig('poly_fit.eps', format='eps')   # save figure to current folder (place before plt.show()).
plt.savefig('R dependence diagram.png', format='png')   # save figure to current folder (place before plt.show()).
plt.show()
```
